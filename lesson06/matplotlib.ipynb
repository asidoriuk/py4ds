```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]

plt.plot(x, y);
```

```python
plt.scatter(x, y);
```

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.title(' f(t)')
plt.xlabel(' t')
plt.ylabel(' f')
plt.axis([-0.5, 9.5, -2.5, 2.5])

plt.plot(t, f, color='Green');
```

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title(' y1')

ax2.plot(x, y2)
ax2.set_title(' y2')

ax3.plot(x, y3)
ax3.set_title(' y3')

ax4.plot(x, y4)
ax4.set_title(' y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.axis([-5, 5, -1, 8]);
```

```python
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv', sep = ',')
df.head()
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
df.Class.value_counts().plot(kind='bar', logy=True);
```

```python
class0 = df.loc[df['Class'] == 0]
class1 = df.loc[df['Class'] == 1]

plt.hist(class0['V1'], bins=20, edgecolor='white', density=True, alpha=.5, label='Class0');
plt.hist(class1['V1'], bins=20, edgecolor='white', density=True, alpha=.5, label='Class1');

plt.xlabel('Class')
plt.ylabel('Density')
plt.legend(loc='best');
```
